## 풀이 코드

### 입력
* 첫째 줄 : 전체 사람의 수 (N) = 노드의 수  



### 출력
* 둘째 줄에서 입력 받은 두 사람의 촌수를 나나태는 정수 반환 (친척관계가 없는 사람들은 -1 을 출력)


### 문제 접근
여러 노드들간의 관계가 주어졌을때 한 노드에서 한 노드까지의 거리를 dfs 로 구현하는 문제
* 여러 사람들 간의 관계를 그래프로 표현이 가능.  


### 문제 해설
문제에서 제시해준 힌트들을 그래프에 대입하여 생각해보면 다음과 같다.   

* 전체 사람의 수는 N 이다 == N 개의 노드를 가진 그래프 


> N개의 노드와, M개의 간선을 가진 그래프가 주어질 때, Target1 노드와 Target2 노드의 거리를 구하는 문제 (단, 서로 다른 그래프일 경우 -1 을 반환하면 됨.)

1. 노드수, 간선 개수 입력받기

````python
N = int(input())
apple = int(input())

apple_loc = []
for _ in range(apple):
    x, y = map(int, input().split())
    apple_loc.append([x, y])

# 뱀의 방향 전환 횟수
plan_cnt = int(input())

# 뱀의 방향 전환 정보
turn_plan = {}
for _ in range(plan_cnt):
    sec, dir_plan = input().split()
    turn_plan[int(sec)] = dir_plan
    
# 북동남서
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

direction = 1

# 방향 전환 함수
def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3
        

def turn_right():
    global direction
    direction += 1
    if direction == 4:
        direction = 0

# 뱀이 이동하는 위치를 저장해야함.
bamm = [[1, 1]]
time = 0
x, y = 1, 1
# 게임 수행
while True:
    time += 1
    
    x = x + dx[direction]
    y = y + dy[direction]
    
    print(x, y, time)
    if (min(x, y) < 1) or (max(x, y) > N) or ([x, y]in bamm):
        break   
        
    if [x, y] in apple_loc:
        bamm.append([x, y])
        apple_loc.remove([x, y])
    else:
        bamm.append([x, y])
        bamm.pop(0)
        
    if time in turn_plan.keys():
        if turn_plan[time] == 'L':
            turn_left()
        else:
            turn_right()

print(time)
````

````python
from collections import deque
N = int(input())
apple = int(input())

apple_loc = []
for _ in range(apple):
    x, y = map(int, input().split())
    apple_loc.append([x, y])

plan_cnt = int(input())

turn_plan = {}
for _ in range(plan_cnt):
    sec, dir_plan = input().split()
    turn_plan[int(sec)] = dir_plan
    

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

direction = 1


def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3
        

def turn_right():
    global direction
    direction += 1
    if direction == 4:
        direction = 0


bamm = deque([[1, 1]])
time = 0
x, y = 1, 1

while True:
    time += 1
    
    x = x + dx[direction]
    y = y + dy[direction]
    
    print(x, y, time)
    if (min(x, y) < 1) or (max(x, y) > N) or ([x, y] in bamm):
        break   
        
    if [x, y] in apple_loc:
        bamm.append([x, y])
        apple_loc.remove([x, y])
    else:
        bamm.append([x, y])
        bamm.popleft()
        
    if time in turn_plan.keys():
        if turn_plan[time] == 'L':
            turn_left()
        else:
            turn_right()

print(time)
````
