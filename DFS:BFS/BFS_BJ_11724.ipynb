import sys
from collections import deque

N, M = map(int, sys.stdin.readline().split())
graph = [[]*(N) for _ in range(N+1)]
visited = [False] * (N+1)
count = 0
                     
def bfs():
    global count
    queue = deque()
    for i in range(1, N+1):
        if not visited[i]:
        # í˜„ì¬ ë…¸ë“œ ë°©ë¬¸ ì²˜ë¦¬ ë° queue ì— ì¶”ê°€
            visited[i] = True
            queue.append(i)
            count += 1
            # íê°€ í…… ë¹Œë•Œê¹Œì§€ ë°˜ë³µ
            while queue:
                v = queue.popleft()

                for node in graph[v]:

                    if not visited[node]:

                        visited[node] = True
    print(count)

    node1, node2 = map(int, sys.stdin.readline().split())
    graph[node1].append(node2)

bfs()
