## 백준온라인 저지 11724 번 문제 : 연결요소의 개수
문제 출처 : https://www.acmicpc.net/problem/11724

방향이 없는 그래프가 주어졌을때 연결요소(Connected Componet)의 개수를 구하는 문제  
* 연결요소란? : 그래프는 여러개의 고립된 부분 그래프로 구성될 수 도 있는데, 이러한 그래프 각각을 연결요소라고 함.  

* 연결요소의 특징
    * 한 연결요소에는 속한 모든 정점을 연결하는 경로가 있어야 함.
    * 다른 연결 요소에 속한 정점과 연결되는 경로가 없어야 함. 
    
 

## 백준 제출 코드
```python
import sys
from collections import deque

N, M = map(int, sys.stdin.readline().split())
graph = [[]*(N) for _ in range(N+1)]
visited = [False] * (N+1)
count = 0
                     
def bfs():
    global count
    queue = deque()
    for i in range(1, N+1):
        if not visited[i]:
        # 현재 노드 방문 처리 및 queue 에 추가
            visited[i] = True
            queue.append(i)
            count += 1
            # 큐가 텅 빌때까지 반복
            while queue:
                v = queue.popleft()
                # 나온 정점의 인접한 노드들을 반복
                for node in graph[v]:
                    # 해당 노드가 방문된 적 없다면
                    if not visited[node]:
                        queue.append(node)
                        visited[node] = True
    print(count)

for _ in range(M):
    node1, node2 = map(int, sys.stdin.readline().split())
    graph[node1].append(node2)
    graph[node2].append(node1)

bfs()
```

## 풀이 코드
1. 첫째 줄의 노드의 수와 간선의 수를 N, M 의 변수값으로 입력받고,  
2. 노드의 개수 + 1(편의상 python 은 0부터 시작하므로) 만큼 2차원 리스트인 graph 를 생성
3. 각 노드의 방문 여부를 확인해줄 visited 변수 생성, N+1 만큼(편의상 python 은 0부터 시작하므로)  

>bfs 로 문제를 풀이하였음. 방향이 없고 그래프의 연결요소를 구하는 문제이므로, 연결요소는 연결요소 안에 속한 노드가 모두 연결되어 있다는 전제조건이 있음. 한 노드에 연결된 모든 노드들을 너비 우선 탐색으로 탐색을 마쳤을 때 연결요소의 개수를 뜻하는 `count` 값에 1을 더하는 방식으로 문제를 풀이함. 

* `def bfs()` 는 입력받은 노드들을 돌아가며 모두 방문. (1)
* 현재 노드를 **방문처리**하고 **queue 에 삽입**. (2)
* queue 의 첫번째 요소(노드) 를 꺼내어 해당 노드와 연결된 노드들이 **방문한적이 없다면**, **방문처리**하고 **queue 에 삽입** (3)
* 위 방법(3)을 queue 의 요소가 하나도 남지 않을 때 까지 반복. (4)

````python
import sys
from collections import deque
# 재귀함수를 사용하고자 할때는 다음 코드를 추가해주어 Recursion limit 을 필요한 만큼 늘려줘야 함.
# sys.setrecursionlimit(10*6)

N, M = map(int, input('노드의 개수와 간선의 개수 : ').split())
graph = [[]*(N) for _ in range(N+1)]
visited = [False] * (N+1)
count = 0
                     
def bfs():
    global count
    queue = deque()
    for i in range(1, N+1):
        print('\n\n 현재 노드 : ', i)
        print(' 노드 방문 상황 : ', visited[1:])
        print('---------------------------')
        if not visited[i]:
        # 현재 노드 방문 처리 및 queue 에 추가
            visited[i] = True
            print(f' ➡ 현재 노드 {i} queue 삽입')
            print(f' ✓  현재 노드 {i} 방문처리 완료')
            queue.append(i)
            print(f'\n📌 현재 queue : {queue}')
            count += 1
            
            # 큐가 텅 빌때까지 반복
            while queue:
                v = queue.popleft()
                print('\n    ----------------------------------')
                print(f'    ⬅ queue에서 첫번째 노드 꺼내기 : {v}번 노드')
                print('    ----------------------------------')
#                 print(f'    현재 queue : {queue}')
                # 나온 정점의 인접한 노드들을 반복
                for node in graph[v]:
                    
                    if not visited[node]:
                        print(f'    ➡ 노드{node} queue 삽입({v}번 노드와 인접)')
                        queue.append(node)
                        print(f'    ✓ {v}번 노드와 인접한 노드{node} 방문처리')
                        visited[node] = True
                    else:
                        print(f'    ✓{v}번 노드와 인접한 {node}노드는 이미 방문했습니다.')
                
                print(f'\n    📌 현재 queue : {queue}')
    print(count)

for _ in range(M):
    node1, node2 = map(int, input('연결된 노드를 입력해주세요 : ').split())
    graph[node1].append(node2)
    graph[node2].append(node1)

bfs()
print('graph : ', graph)
````
